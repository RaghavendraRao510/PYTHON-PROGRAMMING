# What is Python Programming

# Python

* Python is a **high-level programming language**.
* Its syntax is simple and close to English, making it easy for humans to read and understand.
* Python is an **object-oriented programming language (OOP)**.
* Python is an **interpreted language**, meaning the code is executed **line by line**.
* Python is **case-sensitive**. For example, `APPLE`, `Apple`, and `apple` are treated as three different identifiers.
* Python has a **huge collection of libraries** that simplify programming tasks.
* Some popular Python libraries include:

  * **NumPy** – Numerical computing
  * **Pandas** – Data analysis and manipulation
  * **Matplotlib** – Data visualization
  * **Seaborn** – Statistical data visualization
  * **Django** – Web development framework
  * **Flask** – Lightweight web framework
  * **Scikit-learn** – Machine learning
* Python requires **fewer lines of code** compared to many other programming languages, making development faster and more efficient.

A **variable** stores a value.                                                
You create a variable by giving it a name and assigning a value using `=`.

```python
name = "Raghu"
age = 16

print(name)
print(age)
```

```python
number1=10
number2=20
print(number1+number2)
```

**Variable Naming Rules**                                                
Valid variable names:

- Can contain letters, numbers, and `_`
- Cannot start with a number
- Cannot contain spaces
- Cannot use Python keywords such as `if`, `for`, or `class`
- Variable names are case-sensitive

Example: `age` and `Age` are different variables.


```python
student_name = "Ravi"
student2 = "Kiran"
_total = 100

print(student_name)
print(student2)
print(_total)
```

**Basic Python Data Types**

Common data types:

- `int` → whole numbers
- `float` → decimal numbers
- `str` → text
- `bool` → `True` or `False`


```python
age = 16
height = 5.7
name = "Raghu"
is_student = True

print(age)
print(height)
print(name)
print(is_student)
```

**Checking Data Types**

Use `type()` to find the data type of a value or variable.

```python
age = 16
height = 5.7
name = "Python"
is_easy = True

print(type(age))
print(type(height))
print(type(name))
print(type(is_easy))
```

**Simple Practice Program**
